```python
# Import libraries
import pandas as pd

def data_input(csv) -> pd.DataFrame:
    return pd.read_csv(csv)
    
def data_info(data:pd.DataFrame) -> pd.DataFrame:
    return data.info()
    
def data_describe(data) -> pd.DataFrame:
    return data.describe().T
    
def data_isnull(data) -> pd.DataFrame:
    return data.isnull().sum().to_frame().rename(columns={0: 'null_count'})

def data_type(data):
    data = data.dtypes
    return data

def data_count(data):
    count_dict = {}
    for column in data.columns:
        count_dict[column] = [data.shape[0]]
    df = pd.DataFrame(count_dict).T.rename(columns={0: 'entry_count'})
    df['data_type'] = data_type(data)
    return df
```
