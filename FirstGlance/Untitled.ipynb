```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()
num_columns = 10
num_rows = 1000
```

```python
num_creation = fake.random_number(digits=random.randint(4,5))
```

```python
data = {}
for x in range(num_columns):
    column_name = x
    data[column_name] = [fake.random_number(digits=5) for x in range(num_rows)]
```

```python
dataframe = pd.DataFrame(data)
```

```python
dataframe.to_csv('example.csv', index=False)
```
